# Vibration in a one-dimensional system

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def M(w,k):
    j=np.zeros((40,40))
    m=1
    d=(2*k)-(m*(w**2))
    z=d-k
    for i in range(40):
        if i == 0:
            j[i][i]=z
            j[i][i+1]=-k
        elif i == 39:
            j[i][i]=z
            j[i][i-1]=-k
        else :
            j[i][i-1]=-k
            j[i][i]=d
            j[i][i+1]=-k 
    return j
              
def b1(c):
    k=np.zeros((40,1))
    k[0]=c
    return k
```

```python
def graph1():
    y=np.linalg.solve(M(1,1),b1(1)) # w=1일때
    y1=np.linalg.solve(M(2,1),b1(1)) # w=2일때
    y2=np.linalg.solve(M(3,1),b1(1)) # w=3일때
                       
    x=[i for i in range(40)]
    plt.plot(x,y,label='w=1')
    plt.plot(x,y1,label='w=2')
    plt.plot(x,y2,label='w=3')
    plt.xlabel('i')
    plt.ylabel('a_i')
    plt.legend()
    plt.show
```

```python
graph1()
```

### 그래프 의미
w=1, w=3은 역행렬이 존재하지 않아 구할수가 없다.
w=2를 보면 힘이 작용하는 방향으로 부터 멀어질수록 진폭의 크기가 작아져 0에 가까워진다.

#  2.양쪽 끝에서 힘을 가하면 어떻게 되는가?

```python
def b2(c):
    k=np.zeros((40,1))
    k[0]=c
    k[39]=c
    return k
```

```python
def graph2():
    y2=np.linalg.solve(M(1,1),b2(1)) # w=1일때
    y21=np.linalg.solve(M(2,1),b2(1)) # w=2일때
    y22=np.linalg.solve(M(3,1),b2(1)) # w=3일때
                       
    x2=[i for i in range(40)]
    plt.plot(x2,y2,label='w=1')
    plt.plot(x2,y21,label='w=2')
    plt.plot(x2,y22,label='w=3')
    plt.xlabel('i')
    plt.ylabel('a_i')
    plt.legend()
    plt.show
```

```python
graph2()
```

##### 그래프 의미
w=1,w=3에서는 역행렬이 존재하지 않았으며 w=2일 때 그래프를 보면 양쪽에서 진폭이 점점 작아지다가 가운데서 0에 가까워진다.
힘이 양쪽에 있으므로 두 힘이 반대 방향으로 힘이 작용하여 결국 서로 상쇄되어가는 모습이다.
