# Random walk

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
N = 100
T = 50 
 
init_position = np.zeros((N, T+1))
 
for t in range(1, T+1):
    for i in range(N):
        step=[1,-1]
        probabilities=[0.5,0.5]
        init_position[i][t] = init_position[i][t-1] + np.random.choice(step,p=probabilities)
        
plt.plot(init_position[i]);
```

```python
for i in range(N):
    plt.plot(init_position[i])
```

```python
def random_walks(t):
    step = 1, -1
    probabilities = 0.5, 0.5
    x=0
    for i in range(1,t+1):
        y=np.random.choice(step, p=probabilities)
        x=x+y
    return x

N = 10000 
t = 50 
path = []
for i in range(N):
    i = i+1
    path.append(random_walks(t))

x = np.linspace(-50,50,101)
y = []   
z=0
for i in range(-51,50,1):
    i = i+1
    y.append(path.count(i)/N)
    z += path.count(i)/N
    
print(z)
plt.plot(x,y);
```

```python
from sympy import *

t = symbols('t')
r = symbols('r')
x = symbols('x', cls=Function)
```

```python
#r= -1, x(0)=2
deq = Eq(x(t).diff(t),-1*x(t)-x(t)*x(t))
psol1 = dsolve( deq, ics= {x(0):2} )
psol1
```

```python
#r= -1, x(0)=-2
deq = Eq(x(t).diff(t),-1*x(t)-x(t)*x(t))
psol2 = dsolve( deq, ics= {x(0):-2} )
psol2
```

```python
#r= 1, x(0)=2
deq = Eq(x(t).diff(t),1*x(t)-x(t)*x(t))
psol3 = dsolve( deq, ics= {x(0):2} )
psol3
```

```python
#r= 1, x(0)=-2
deq = Eq(x(t).diff(t),1*x(t)-x(t)*x(t))
psol4 = dsolve( deq, ics= {x(0):-2} )
psol4
```

```python
plot(psol1.rhs, (t,-1,1))
plot(psol2.rhs, (t,-1,1))
plot(psol3.rhs, (t,-1,1))
plot(psol4.rhs, (t,-1,1));
```
