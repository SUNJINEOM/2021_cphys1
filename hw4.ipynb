### Error Function

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import math
from scipy.misc import derivative
```

```python
def erf(x):
    return 2*(np.pi**(-1/2))*(np.exp(-x**2))
```

```python
def integ_erf(erf,a,b,n):
    s, x, dx = -1, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*erf(x)+0.5*erf(x+dx))
        x = x+dx
    return s
```

```python
x = np.arange(-4,4,0.01)
plt.axis([-4,4,-1.5,1.5])
plt.plot(x,[integ_erf(erf,-5,k,500) for k in x])
plt.title('erf')
plt.grid()
plt.show()
```

```python
integrate.quad(erf,0,6)
```

### Error Function Talyor Expend

```python
def erf_taylor(x,n):
    f=0
    for i in range(n+1):
        f=f+(2/np.pi**(0.5))*(((-1)**i)*(x**(2*i+1)))/((2*i+1)*math.factorial(i))
    return f
```

```python
plt.plot(x,[integ_erf(erf,-5,k,500) for k in x], label='erf')
plt.plot(x, erf_taylor(x, 1), label='erf_taylor n=1',color='purple')
plt.plot(x, erf_taylor(x, 9), label='erf_taylor n=9',color='red')
plt.plot(x, erf_taylor(x, 10), label='erf_taylor n=10',color='green')
plt.plot(x, erf_taylor(x, 50), label='erf_taylor n=50',color='blue')
plt.plot(x, erf_taylor(x, 100), label='erf_taylor n=100',color='black')
plt.grid()
plt.legend()
plt.title('Error_Function')
plt.axis([-5,5,-2,2])
plt.show()
```

### scipy.Integrate

```python
def sin(x):
    return np.sin(x)
integrate.quad(sin,0,6)
```

```python
def cos(x):
    return np.cos(x)
integrate.quad(cos,0,6)
```

```python
def sin(t):
    return np.sin(t)

def cos(t):
    return np.cos(t)
    

def sincos():
    x = np.arange(0,6,0.01)
    u,v = [],[]
    
    for i in x :
        f,s = integrate.quad(sin,0,i)
        g,t = integrate.quad(cos,0,i)
        u.append(f)
        v.append(g)
    plt.title("Integration")
    plt.plot(x,u,label='sin_Integration')
    plt.plot(x,v,label='cos_Integration')
    plt.grid()
    plt.legend()
    plt.show()
    
sincos()
```
